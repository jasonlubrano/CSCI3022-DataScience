Let $x_1, x_2, \ldots, x_n$ be $n$ observations of a variable of interest.  Recall that the sample mean $\bar{x}_n$ and sample variance $s^2_n$ are given by 

$$
\bar{x}_n = \frac{1}{n}\sum_{k=1}^n x_k \quad \textrm{and} \quad s^2_n = \frac{1}{n-1}\sum_{k=1}^n \left( x_k - \bar{x}_n\right)^2 \qquad \tag{Equation 1}
$$

where here the subscript $n$'s indicate the number of observations in the sample. Notice that a natural computation of the variance requires two passes over the data: one to compute the mean, and a second to subtract the mean from each observation and compute the sum of squares. It is often useful to be able to compute the variance in a single pass, inspecting each value $x_k$ only once; for example, when the data are being collected without enough storage to keep all the values, or when costs of memory access dominate those of computation. In this problem you will explore two methods for such an _online_ computation of the mean.  

**Part A**: Show algebraically that the following relation holds between the mean of the first $n-1$ observations and the mean of all $n$ observations: 

$$
\bar{x}_n = \bar{x}_{n-1} + \frac{x_n - \bar{x}_{n-1}}{n}
$$

Note that you can get an expression for $\bar{x}_{n-1}$ by simply replacing $n$ in Equation 1 above with $n-1$.

\begin{align}
\bar{x}_{n} &= \bar{x}_{n-1}+\frac{x_{n}-\bar{x}_{n-1}}{n}\\
    &= \frac{n\bar{x}_{n-1} + x_{n} - \bar{x}_{n-1}}{n} \\
    &= \frac{n\bar{x}_{n-1} - \bar{x}_{n-1} + x_{n}}{n} \\ 
    &= \frac{\bar{x}_{n-1}(n-1) + x_{n}}{n} \\
n\bar{x}_{n} &= \bar{x}_{n-1}(n-1) + x_{n} \\
n\bar{x}_{n} - x_n &= \bar{x}_{n-1}(n-1) \\
x_{n}(n - 1) &= \bar{x}_{n-1}(n-1) \\
x_{n} &= \bar{x}_{n-1}
\end{align}

**Part C:** One might hypothesize that students' grades are higher in one credit-hour courses than in regular lecture sections. Note that `Hours` corresponds to the number of credit-hours a class is worth, and that Average Grades are provided in the data set on a GPA-scale (4=A, 3=B, and so on).

Compute the mean and median Average Grade for all undergraduate Computer Science courses, and for only the one credit-hour undergraduate Computer Science courses. Report all four of these numbers, and comment on the results. Does your calculation seem to support our hypothesis or not? 

```python
lec_dfC = dfC[(dfC['Activity_Type'] == "LEC - Lecture")] #bc the instructions specifically said LECTURE SECTIONS
one_dfC = dfC[(dfC['Hours'] == 1)]
thr_dfC = dfC[(dfC['Hours'] == 3) & (dfC['Activity_Type'] == "LEC - Lecture")]
fou_dfC = dfC[(dfC['Hours'] == 4) & (dfC['Activity_Type'] == "LEC - Lecture")]
#made four data sets for each hour class

# all undergraduate average
all_cred_tot = lec_dfC['AVG_GRD'].sum()
#print(all_cred_tot)
all_cred_num = lec_dfC['AVG_GRD'].count()
#print(all_cred_num)
all_cred_avg = all_cred_tot / all_cred_num
print('all credit hour undergraduate average:', all_cred_avg)

# all undergraduate median
all_cred_med = lec_dfC['AVG_GRD'].sort_values().median()
print('all credit hour undergraduate median:', all_cred_med)

# one credit hour undergraduate average
one_cred_tot = one_dfC['AVG_GRD'].sum()
#print(one_cred_tot)
one_cred_num = one_dfC['AVG_GRD'].count()
#print(one_cred_num)
one_cred_avg = one_cred_tot / one_cred_num
print('one credit hour undergraduate average:', one_cred_avg)

# one credit hour undergraduate median
one_cred_med = one_dfC['AVG_GRD'].sort_values().median()
print('one credit hour undergraduate median:', one_cred_med)

# three credit hour undergraduate average
thr_cred_tot = thr_dfC['AVG_GRD'].sum()
#print(thr_cred_tot)
thr_cred_num = thr_dfC['AVG_GRD'].count()
#print(thr_cred_num)
thr_cred_avg = thr_cred_tot / thr_cred_num
print('three credit hour undergraduate average:', thr_cred_avg)

# three credit hour undergraduate median
thr_cred_med = thr_dfC['AVG_GRD'].sort_values().median()
print('three credit hour undergraduate median:', thr_cred_med)

# four credit hour undergraduate average
fou_cred_tot = fou_dfC['AVG_GRD'].sum()
#print(fou_cred_tot)
fou_cred_num = fou_dfC['AVG_GRD'].count()
#print(fou_cred_num)
fou_cred_avg = fou_cred_tot / fou_cred_num
print('four credit hour undergraduate average:', fou_cred_avg)


# four credit hour undergraduate median
fou_cred_med = fou_dfC['AVG_GRD'].sort_values().median()
print('four credit hour undergraduate median:', fou_cred_med)

```
